### 1.2 유저 데이터 입력

```python
user = [6,1,8,4,8,1,0,1,0,1,1,0,1,1]
# user1 = [8,1,6,9,6,0,0,1,0,0,0,1,0,1]
```

```python
def user_transform(user): # min_Max Scaling 함수
    user_min = min(user)
    user_max = max(user)
    
    user_scaled = []
    for u in user:
        x = (u - user_min) / (user_max - user_min)
        user_scaled.append(x)
    return user_scaled
```

```python
user_scaled = user_transform(user)
user_scaled
```

```python
def weighting(user_data, weight): # 유저 선호도 데이터 : n-array, 가중치 데이터:n x n matrix
    user_weighted = []
    for i in range(len(weight)):
        each_weighted = []
        for k in range(len(weight)):
            y = user_data[i] * weight[i][k]
            each_weighted.append(y)
        user_weighted.append(each_weighted)
    return user_weighted
```

## 3. 유사도 판정

```python
def similarity(user_df, df, user_name, num): # 유저 데이터, 유사도 측정을 위한 데이터, 유저 이름, 원하는 순위
    con_data = pd.concat([user_df.loc[[user_name]],df])
    rc_sim = cosine_similarity(con_data,con_data)
    sim_matrix = pd.DataFrame(rc_sim,columns=con_data.index).loc[[user_name]].T
    rank = sim_matrix[user_name].sort_values(ascending=False) # 유사도 순서로 정렬
    ranking = rank[1:num+1].index.tolist() # 1~n 위 리스트
    return ranking
```
