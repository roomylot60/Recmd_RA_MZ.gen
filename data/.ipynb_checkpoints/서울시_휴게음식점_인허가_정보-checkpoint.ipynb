#### 편의점 데이터 프레임 및 csv 파일 생성

```python
conv_df = df[df['업태구분명']=='편의점']
```

```python
conv_df.to_csv('Convenience_Store_in_Seoul')
```

#### 카페, 다방 분류를 통합하여 데이터 프레임으로 형성 후 csv 파일로 변환

```python
cafe_df = df[df['업태구분명']=='커피숍']
```

```python
# 카페 중 스타벅스만 추출해서 csv 파일로 변환
starbucks = cafe_df[cafe_df['사업장명'].str.contains('스타벅스')]
starbucks.to_csv('Starbucks_in_Seoul.csv',mode='w')
```

```python
cof_df = df[df['업태구분명']=='다방']
```

```python
cafe_df['사업장명'].value_counts()
```

```python
cof_df['사업장명'].value_counts()
```

```python
cafe_df = pd.concat([cafe_df,cof_df])
```

```python
cafe_df.to_csv('Cafe_in_Seoul.csv',mode='w')
```
