```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('tmp_data.csv')
df
```

```python
df.rename(columns={'Unnamed: 0': '동이름'},inplace=True)
```

```python
df
```

```python
df = df.set_index('동이름')
df
```

```python
kmeans = KMeans(n_clusters=10, init='k-means++',max_iter=300,random_state=0)
kmeans.fit(df)
```

```python
print(kmeans.labels_)
```

```python
df['cluster'] = kmeans.labels_ # 군집화 분류값
```

```python
df
```

```python
df[df['cluster']==3]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_transformed = pca.fit_transform(df)
df['pca_x']=pca_transformed[:,0]
df['pca_y']=pca_transformed[:,1]
```

```python
df
```

```python
plt.scatter(x=df.loc[:,'pca_x'], y=df.loc[:,'pca_y'],c=df['cluster'])
```
