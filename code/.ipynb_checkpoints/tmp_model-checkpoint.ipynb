## 임시 모델 작성

### 1.1 import packages and modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cluster import KMeans
import warnings

warnings.filterwarnings(action = 'ignore')
```

### 1.2 Read data

```python
df = pd.read_csv('tmp_data.csv')
df
```

```python
df.rename(columns={'Unnamed: 0':'동이름'},inplace=True)
```

```python
df.set_index('동이름')
```

```python
idx = df['동이름']
```

```python
data = df[['A','B','C','D']].values
```

### 1.3 K-means clustering

```python
kmeans = KMeans(n_clusters=10, init='k-means++',max_iter=300,random_state=0)
kmeans.fit(data)
```

```python
kmeans.labels_
```

```python
tmp_df = df.copy()
```

```python
tmp_df['km_cluster'] = kmeans.labels_
tmp_df.set_index('동이름',inplace=True)
```

```python
tmp_df.groupby('km_cluster').count()
```

### 1.4 유저 데이터 입력

```python
user = [1,6,2,2]
```

```python
user1 = [8,4,5,6]
```

```python
user2 = [4,8,6,3]
```

```python
user3 = [1,3,5,3]
```

**유저 데이터 가공을 통해 군집화에 적합한 형태로 변형**
- 임시로 10으로 나눈 값을 사용

```python
user_td = [x/10 for x in user]
user_td
```

```python
user_sum = [user, user1, user2, user3]
```

```python
user_name = ['user', 'user1', 'user2', 'user3']
```

```python
user_all = []
for u in user_sum:
    user_tmp = [x/10 for x in u]
    user_all.append(user_tmp)
user_all
```

#### 2.1 유저 데이터 예측

```python
user_df = pd.DataFrame(user_all,columns=['A','B','C','D'],index=user_name)
```

```python
user_df
```

```python
cluster_user = kmeans.predict(user_df)
```

```python
user_df['km_cluster'] = cluster_user
```

```python
user_df
```

### 2.2 지역 추출

```python
tmp_df['km_cluster']
```

```python
rc_seoul = tmp_df[tmp_df['km_cluster'] == rc_area]
rc_seoul # 추천 서울 동
```
