### 1.1 import packages and modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cluster import KMeans
import warnings

warnings.filterwarnings(action = 'ignore')
```

### 1.2 Read data

```python
df = pd.read_csv('tmp_data.csv')
df
```

```python
df.rename(columns={'Unnamed: 0':'동이름'},inplace=True)
```

```python
df.set_index('동이름')
```

```python
idx = df['동이름']
```

```python
data = df[['A','B','C','D']].values
```

### 1.3 K-means clustering

```python
kmeans = KMeans(n_clusters=10, init='k-means++',max_iter=300,random_state=0)
kmeans.fit(data)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
tmp_df = df.copy()
```

```python
tmp_df['km_cluster'] = kmeans.labels_
tmp_df.set_index('동이름',inplace=True)
```

```python
tmp_df.groupby('km_cluster').count()
```

### 1.4 유저 데이터 입력

```python
user = [1,6,2,2]
```

```python
user1 = [8,4,5,6]
```

```python
user2 = [4,8,6,3]
```

**유저 데이터 가공을 통해 군집화에 적합한 형태로 변형**
- 임시로 10으로 나눈 값을 사용

```python
user_td = [x/10 for x in user]
user_td
```

```python
user_sum = [user, user1, user2, user3]
```

```python
user_name = ['user', 'user1', 'user2', 'user3']
```

### 3. Cosine 유사도를 통해 유저 데이터와 비교
- cosine 유사도에서는 각 feature를 벡터 요소로 판단
- 각 feature에 대한 크기와 관계없이 전체적인 벡터의 방향을 비교하게 되어 비율의 유사도는 측정할 수 잇을 듯
- 어느 한 feature에 대한 가중치가 큰 것을 요구할 경우에 이를 반영함에 있어서 오류가 있을 수도 있을 것 같다

```python
from sklearn.metrics.pairwise import cosine_similarity

con_data = pd.concat([user_df.loc[['user']],rc_seoul])
con_data
```

```python
rc_sim = cosine_similarity(con_data,con_data) # 유사도 행렬
```

```python
sim_matrix = pd.DataFrame(rc_sim,columns=con_data.index).loc[[0]].T
```

```python
rank = sim_matrix[0].sort_values(ascending=False) # 유사도 순서로 정렬
```

```python
rank[1:4].index.tolist() # 1~3 위
```

### 4. 유사도 측정

- Enclidean
- Manhattan
- Standardized
- Mahalanobis

```python
def euclidean_similarity(x,y):
    distance = np.sqrt(sum((x-y)**2))
    return distance
```

```python
user_df.loc['user']
```

```python
rc_seoul.iloc[0]
```

```python
euclidean_similarity(user_df.loc['user'],rc_seoul.iloc[0])
```

```python
eu_sim = []
for i in range(len(rc_seoul)):
    eu_sim.append(euclidean_similarity(user_df.loc['user'],rc_seoul.iloc[i]))
    
eu_df = pd.DataFrame(eu_sim, index=rc_list)
```
