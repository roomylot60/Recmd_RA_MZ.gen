```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup

import pandas as pd
import numpy as np
import time

ser = Service('chromedriver.exe')
driver = webdriver.Chrome(service = ser)

url = 'https://www.instagram.com/'
driver.get(url)
```

### 인스타그램 창이 뜨면 로그인 진행 후 아래의 함수들을 실행시켜주세요!

```python
def insta_searching(word):
    url = 'https://www.instagram.com/explore/tags/{}/'.format(word)
    return(url)

from selenium.webdriver.common.by import By

def select_first(driver):
    # 웹페이지에 접속해서 첫번째 게시글 해당하는 tag 선택해줘
    first = driver.find_element(By.CSS_SELECTOR, 'div._aagu')

    first.click()
    time.sleep(3)

import re
def get_content(driver):
    html_insta = driver.page_source
    soup = BeautifulSoup(html_insta, 'html.parser')

    try:
        content = soup.select('div._a9zs > span ')[0].text
        print(content)

    except:
        content = " "

    try:
        tags = re.findall(r'#[^\s#,\\]+',content)
    except:
        tags = " "

    data = [content, tags]
    print(data ,"!!!")
    return(data)

def move_next(driver):
    right = driver.find_element(By.CSS_SELECTOR,"div._aaqg")

    right.click()
    time.sleep(3)

def insta_crawling(word, n):

    url = insta_searching(word)
    driver.get(url)
    time.sleep(8)


    # 첫 사진 선택
    select_first(driver)
    time.sleep(8)


    results = []
    for i in range(n):
        time.sleep(2)
        try:
            data = get_content(driver)
            results.append(data)
            move_next(driver)

        except:
            time.sleep(2)
            move_next(driver)

    return(results)
```

### tag 변수에 크롤링 할 태그 입력

```python
tag = "쏘카"
results = insta_crawling(tag, 100)
results_df = pd.DataFrame(results)
results_df.columns = ['content','tags']
```

### 크롤링한 데이터 파일에 저장
- tag 이름 과 현재 시간으로 파일 이름 자동 설정 됩니다 !
- 파일 2개 저장됨
    - 태그 크롤링한 파일 (쏘카_202210211600_tag.txt)
    - 본문 내용 크롤링한 파일 (쏘카_202210211600_content.txt)

```python
from datetime import datetime

now = datetime.now().strftime("%Y%m%d%H%M")

f = open(f'{tag}_{now}_tag.txt','w',encoding='UTF-8')

all_tag_str = ""
for item in results_df['tags'].values:
    str = ' '.join(item)
    all_tag_str += str
    
f.write(all_tag_str)

f.close()

### 관련 없는 태그 삭제

remove_tags = ['육아','맘','딸','기록','아들','일기','도치','아기','선','소띠','호랑띠',\
               '좋아요','환영','소통','맞팔','좋반','선팔','일상','stagram','insta','협찬','그램','데일리','daily','스타그램',\
                '맞','많','너무','우리','더','맞']

for n in remove_tags:
    all_tag_str = all_tag_str.replace(n,'',-1)
```

- 워드클라우드 확인
- 의미 없는 단어 많을 경우 위의 코드에서 remove_tags 에 단어 추가
- 다시 실행 후 워드클라우드 재 생성

```python
from wordcloud import WordCloud
import matplotlib.font_manager as fm
path = "c:/Windows/Fonts/malgun.ttf"

now = datetime.now().strftime("%Y%m%d%H%M")

wc = WordCloud(font_path=path, width=600, height=600,background_color='black',colormap='hsv')
wc.generate(all_tag_str)


wc.to_file(f'{tag}_{now}_tag.png')
```

### 본문 내용 워드 클라우드

```python
now = datetime.now().strftime("%Y%m%d%H%M")

f = open(f'{tag}_{now}_content.txt','w',encoding='UTF-8')

all_content_str = ""
for item in results_df['content'].values:
    str = ' '.join(item)
    all_content_str += str
    
f.write(all_content_str)

f.close()

remove_tags = ['육아','맘','딸','기록','아들','일기','도치','아기','선','소띠','호랑띠',\
               '좋아요','환영','소통','맞팔','좋반','선팔','일상','stagram','insta','협찬','그램','데일리','daily','스타그램',\
                '맞','많','너무','우리','더','맞']

## 컨텐츠 긁어오면 띄어쓰기가 이상해서
# 그거 제거 하는 코드입니다!
# 크게 신경쓰지 않으셔도 됩니당 : )
all_content_str = all_content_str.replace("   ",'***',-1)
all_content_str = all_content_str.replace(" ",'',-1)
all_content_str = all_content_str.replace("***", " ", -1)

for n in remove_tags:
    all_content_str = all_content_str.replace(n,'',-1)
```

```python
from wordcloud import WordCloud
import matplotlib.font_manager as fm
path = "c:/Windows/Fonts/malgun.ttf"

now = datetime.now().strftime("%Y%m%d%H%M")

wc = WordCloud(font_path=path, width=600, height=600,background_color='black',colormap='hsv')
wc.generate(all_content_str)


wc.to_file(f'{tag}_{now}_content.png')
```
